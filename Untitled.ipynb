```python
data_sorted=data[data['title'].apply(lambda x: len(x.split())>4)]
```

```python
data_sorted.shape
```

```python
data_sorted.sort_values('title',inplace=True,ascending=False)
```

```python
data_sorted.head()
```

```python
import itertools
num_examples=data_sorted.shape[0]
i=0
j=0
stage1_dedupe_asins=[]
num_data_points=num_examples
while i<num_examples:
    previous_i=i
    a=data_sorted['title'].iloc[i].split()
    
    j=i+1
    while j<num_examples:
        
        b=data_sorted['title'].iloc[j].split()
        length=max(len(a),len(b))
        count=0
        for k in itertools.zip_longest(a,b):
            if(k[0]==k[1]):
                count+=1
        if(length-count)>2:
            stage1_dedupe_asins.append(data_sorted['asin'].iloc[i])
            if j == num_data_points-1: stage1_dedupe_asins.append(data_sorted['asin'].iloc[j])
            i = j
            break
        else:
            j += 1
    if previous_i == i:
        break    
            
```

```python
data = data.loc[data['asin'].isin(stage1_dedupe_asins)]
```

```python
stop_words=set(stopwords.words('english'))
```

```python
def nlp_preprocessing(total_text,index,column):
    if type(total_text) is not int:
        string=""
        for words in total_text.split():
            word=("".join(e for e in words if e.isalnum()))
            word=word.lower()
            if not word in stop_words:
                string+=word+" "
        data[column][index]=string
```
